```python
class RepeatReplacer(object):
  def __init__(self):
    self.repeat_regexp = re.compile(r"(\w*)(\w)\2(\w*)")
    self.repl = r"\1\2\3"
  def replace(self, word):
    if (wn.synsets(word)):
      return word
    repl_word = self.repeat_regexp.sub(self.repl, word)
    if (repl_word != word):
      return self.replace(repl_word)
    else:
      return repl_word
```

##Most similar words in a list

```python
from nltk.corpus import wordnet

def mostsimilar(my_list):
  max_sim = 0
  most_similar_words = tuple()
  for i in range(0, len(my_list)-1):
    for syn1 in wordnet.synsets(my_list[i]):
      for j in range(i+1, len(my_list)):
        for syn2 in wordnet.synsets(my_list[j]):
          x=syn1.wup_similarity(syn2)
          if(x):
            if(x > max_sim):
              most_similar_words = my_list[i], my_list[j], x
              max_sim = x
  return most_similar_words
```

```python
mostsimilar(["car", 'knee', "food", "fruit", "eye"])
```

```python
def ins_correct(word):
  if(wn.synsets(word)):
    return word
  words = []
  for i in range(0, len(word)+1):
    first = str(word[0:i])
    last = str(word[i:len(word)])
    for alpha in alphabet:
      new_word = first + alpha + last
      if(wn.synsets(new_word)):
        words.append(new_word)
  return (set(words))
```

```python
ins_correct("crrect")
```

```python
def del_correct(word):
  if(wordnet.synsets(word)):
    return word
  words = []
  for i in range(0, len(word)):
    new_word = str(word[0:i]) + str(word[i+1:len(word)])
    if (wordnet.synsets(new_word)):
      words.append(new_word)
  return set(words)
```

## Defining a Mixed Function of taggers

```python
def backoff_tagger(train_sentences, tagger_methods, backoff=None):
  for cls in tagger_methods:
    backoff = cls(train_sentences, backoff=backoff)
  return backoff
```

```python
backoff = DefaultTagger("NN")
tagger5 = backoff_tagger(train_sents, [UnigramTagger, BigramTagger, TrigramTagger], backoff=backoff)
tagger5.evaluate(test_tag)
```

## Bag of words

```python
def bag_of_words(words):
  return dict([(word, True) for word in words])
```

```python
bag_of_words(["the", "quick", "brown", "fox"])
```

```python
def bag_of_words_not_in_use(words, badwords):
  return bag_of_words(set(words)-set(badwords))
```

```python
bag_of_words_not_in_use(["the", "quick", "brown", "fox"], ["the"])
```

```python
# Second way to remove extra/unnecessaary words
from nltk.corpus import stopwords

def bag_of_non_stop_words(words, stopfile="english"):
  badwords = stopwords.words(stopfile)
  return bag_of_words_not_in_use(words, badwords)
```

```python
bag_of_non_stop_words(["the", "quick", "brown", "fox"])
```

## Using Bigram in Bag of Words

```python
from nltk.collocations import BigramCollocationFinder
from nltk.metrics import BigramAssocMeasures

def bag_of_bigrams(words, score_fn=BigramAssocMeasures.chi_sq, n=200): # n: number of Bigrams
  bcf = BigramCollocationFinder.from_words(words)
  bigrams = bcf.nbest(score_fn, n)
  return bag_of_words(words + bigrams)
```

```python
bag_of_bigrams(["the", "quick", "brown", "fox"])
```
